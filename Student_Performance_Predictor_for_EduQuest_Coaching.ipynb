```python
def remove_outliers_iqr(df, cols):
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

df_no_outliers = remove_outliers_iqr(df_encoded_v1, numeric_cols)
```

```python
from sklearn.preprocessing import StandardScaler

sc= StandardScaler()
print(x_train_v1.shape)
x_train_v1_scaled = sc.fit_transform(x_train_v1)
print(x_test_v1.shape)
x_test_v1_scaled = sc.transform(x_test_v1)
```

```python
sc = StandardScaler()
x_train_v2_scaled = sc.fit_transform(x_train_v2)
x_test_v2_scaled = sc.transform(x_test_v2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_model(y_true, y_pred):

    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)

    metrics = {
        'MAE': round(mae, 4),
        'MSE': round(mse, 4),
        'RMSE': round(rmse, 4),
        'RÂ² Score': round(r2, 4)
    }

    for k, v in metrics.items():
        print(f"{k:8}: {v}")
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

linear_model.fit(x_train_v1_scaled, y_train_v1)
```

```python
y_pred= linear_model.predict(x_test_v1_scaled)

evaluate_model(y_test_v1,y_pred)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor


models = {
    'Linear': LinearRegression(),
    'Ridge': Ridge(alpha=1.0),
    'Lasso': Lasso(alpha=0.1),
    'RandomForest': RandomForestRegressor(n_estimators=100, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
}

for name, model in models.items():
    x_train = x_train_v1_scaled if name in ['Linear','Ridge','Lasso'] else x_train_v1
    x_test = x_test_v1_scaled if name in ['Linear','Ridge','Lasso'] else x_test_v1
    model.fit(x_train, y_train_v1)
    y_pred = model.predict(x_test)
    print(f"{name} Model:")
    evaluate_model(y_test_v1, y_pred)
    print("-" * 50)

```

```python
from sklearn.linear_model import LassoCV

lasso_cv = LassoCV(alphas=np.linspace(0.01, 1, 50), cv=5)
lasso_cv.fit(x_train_v1_scaled, y_train_v1)

print("Best alpha:", lasso_cv.alpha_)
```

```python
lasso_fine_tuned = Lasso(alpha =0.11102040816326529)

lasso_fine_tuned.fit(x_train_v1_scaled, y_train_v1)

y_pred = lasso_fine_tuned.predict(x_test_v1_scaled)

evaluate_model(y_test_v1, y_pred)
```

```python
models = {
    'Linear': LinearRegression(),
    'Ridge': Ridge(alpha=1.0),
    'Lasso': Lasso(alpha=0.1),
    'RandomForest': RandomForestRegressor(n_estimators=100, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
}

for name, model in models.items():
    x_train = x_train_v2_scaled if name in ['Linear','Ridge','Lasso'] else x_train_v2
    x_test = x_test_v2_scaled if name in ['Linear','Ridge','Lasso'] else x_test_v2
    model.fit(x_train, y_train_v2)
    y_pred = model.predict(x_test)
    print(f"{name} Model:")
    evaluate_model(y_test_v2, y_pred)
    print("-" * 50)
```

```python

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, mean_squared_error


def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmse_scorer = make_scorer(rmse, greater_is_better=False)

rf = RandomForestRegressor(random_state=42)

rf_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}

rf_grid = GridSearchCV(rf, rf_param_grid, cv=5, scoring=rmse_scorer, n_jobs=-1, verbose=1)
rf_grid.fit(x_train_v2, y_train_v2)

print("\n Best Random Forest Params:")
print(rf_grid.best_params_)

best_rf = rf_grid.best_estimator_
y_pred_rf = best_rf.predict(x_test_v2)

print("\n Random Forest Performance:")
evaluate_model(y_test_v2, y_pred_rf)

xgb = XGBRegressor(random_state=42, verbosity=0)

xgb_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 6, 10],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

xgb_grid = GridSearchCV(xgb, xgb_param_grid, cv=5, scoring=rmse_scorer, n_jobs=-1, verbose=1)
xgb_grid.fit(x_train_v2, y_train_v2)

print("\n Best XGBoost Params:")
print(xgb_grid.best_params_)

best_xgb = xgb_grid.best_estimator_
y_pred_xgb = best_xgb.predict(x_test_v2)

print("\n XGBoost Performance:")
evaluate_model(y_test_v2, y_pred_xgb)
```
